```
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())

sh = gc.create('coin_models_4')
worksheet_config = sh.add_worksheet(title="config", rows="20", cols="20")
worksheet_config.update_cell(1, 1, 'Present row to process: ')
worksheet_config.update_cell(1, 2, 2)

worksheet_models = sh.add_worksheet(title="models", rows="1000", cols="20")

worksheet_models.update_cell(1, 1, 'model_number')
worksheet_models.update_cell(1, 2, 'batch_size')
worksheet_models.update_cell(1, 3, 'filters1')
worksheet_models.update_cell(1, 4, 'dropout1')
worksheet_models.update_cell(1, 5, 'filters2')
worksheet_models.update_cell(1, 6, 'dropout2')
worksheet_models.update_cell(1, 7, 'filters3')
worksheet_models.update_cell(1, 8, 'dropout3')

#column_titles = ['model_number','batch_sizejj','filters1','dropout1','filters2','dropout2','filters3','dropout3']
#cell_list = worksheet_models.range(1,1,1,8)
#worksheet_models.update_cells(cell_list,column_titles )

model_row = 2
for batch_size_counter in range(1, 3):
    for filters1_counter in range(1, 4):
        for dropout1_counter in range(0, 3):
            for filters2_counter in range(1, 4):
                for dropout2_counter in range(0, 3):
                    for filters3_counter in range(1, 4):
                        for dropout3_counter in range(0, 3):
                            worksheet_models.update_cell(model_row, 1, model_row)
                            worksheet_models.update_cell(model_row, 2, batch_size_counter *8)
                            worksheet_models.update_cell(model_row, 3, 2**(filters1_counter+4))
                            worksheet_models.update_cell(model_row, 4, dropout1_counter * 0.25)
                            worksheet_models.update_cell(model_row, 5, 2**(filters2_counter+4))
                            worksheet_models.update_cell(model_row, 6, dropout2_counter*0.25)
                            worksheet_models.update_cell(model_row, 7, 2**(filters3_counter+4))
                            worksheet_models.update_cell(model_row, 8, dropout3_counter*0.25)
                            print(model_row)
                            model_row = model_row +1
                            
                  

  
```
